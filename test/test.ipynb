```python
import numpy as np 
import pandas as pd
```

```python
intent = "" 
sentences = [] 
classes = []

with open('C:/Users/LENOVO/Documents/GitHub/potential-octo-spork/spell-corrector/training-data.txt','r', encoding="utf8") as file: 
    for line in file:     
        arr = line.split()
        if len(arr) == 0 : 
            continue 

        if arr[1] == "intent:":
            intent = arr[2] 
            continue 

        if arr[0] == "examples:":
            continue 

        sentences.append(line) 
        classes.append(intent) 

print(intent) 
print(len(sentences))
print(len(classes))
```

```python
data = {
    "sentences" : sentences, 
    "classes" : classes
}
df = pd.DataFrame(data)
df.head(10)

# correct the sentences
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sentences, classes, test_size=0.1, random_state=42, shuffle=True)
```

```python
X_train_list = list(X_train)
for i in range(len(X_train_list)):
    X_train_list[i] = ' '.join([str(elem) for elem in X_train_list[i].split()[1:]])

intent = "" 
for i in range(len(X_train_list)):
    if y_train[i] != intent : 
        intent = y_train[i] 
        print("\n")
        print("- intent: ", intent) 
        print("  examples: |")
    
    print("    - ", X_train_list[i])
```

```python
X_test_list = list(X_test)
for i in range(len(X_test_list)):
    X_test_list[i] = ' '.join([str(elem) for elem in X_test_list[i].split()[1:]])

intent = "" 
for i in range(len(X_test_list)):
    if y_test[i] != intent : 
        intent = y_test[i] 
        print("\n")
        print("- intent: ", intent) 
        print("  examples: |")
    
    print("    - ", X_test_list[i])
```
